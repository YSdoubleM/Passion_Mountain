#### 성능확인을 위한 함수

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, roc_auc_score)

def get_clf_eval(y_test, pred):
    acc = accuracy_score(y_test, pred)
    pre = precision_score(y_test, pred)
    re = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    auc = roc_auc_score(y_test, pred)
    
    return acc, pre, re, f1, auc
```

```python
from sklearn.metrics import confusion_matrix

def print_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    acc, pre, re, f1, auc = get_clf_eval(y_test, pred)
    
    print('=> confusion matrix')
    print(confusion)
    print('=============')
    
    print('Accuracy: {0:.4f}, Precision: {1:.4f}'.format(acc, pre))
    print('Recall: {0:.4f}, F1: {1:.4f}, AUC:{2:.4f}'.format(re, f1, auc))
```

### 최적의 모델 찾기
- pipeline >> gridsearchcv

```python
import numpy as np
from sklearn.pipeline import Pipeline

pipe = Pipeline([('clf', RandomForestClassifier())])

search_space = [{'clf': [DecisionTreeClassifier()],
                 'clf__max_depth': [3, 5, 10],
                 'clf__random_state': [10, 13, 15]},
                {'clf': [RandomForestClassifier()],
                 'clf__n_estimators': [10, 100, 1000],
                 'clf__max_depth': [3, 5, 10],
                 'clf__random_state': [10, 13, 15]},
                {'clf': [AdaBoostClassifier()],
                 'clf__n_estimators': [30, 100, 1000],
                 'clf__learning_rate': [1.0, 0.5, 0.1],
                 'clf__random_state': [10, 13, 15]},
                {'clf': [GradientBoostingClassifier()],
                 'clf__n_estimators' : [100, 500],
                 'clf__learning_rate' : [0.05, 0.1]},
                {'clf': [XGBClassifier()],
                 'clf__n_estimators': [100, 500, 1000],
                 'clf__learning_rate': [0.05, 0.1, 0.3, 0.6],
                 'clf__max_depth': [3, 5, 10]}]
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(pipe, search_space, cv=5, verbose=0)
best_model = clf.fit(X_train, y_train)
```

```python
def distribute_region(loc):
    if "경상" in loc or"경남" in loc or "경북" in loc:
        return "경상도"
    elif "강원" in loc:
        return "강원도"
    elif "경기" in loc:
        return "경기도"
    elif "서울" in loc:
        return "서울"
    elif "충청" in loc or "충남" in loc or "충북" in loc:
        return "충청도"
    elif "전라" in loc or "전북" in loc or "전남" in loc:
        return "전라도"
    elif "상북도" in loc:
        return "경상도"
    elif "울산" in loc:
        return "경상도"
    elif "횡성" in loc:
        return "강원도"
    elif "대구" in loc:
        return "경상도"
    elif "부산" in loc:
        return "경상도"
    elif "제주" in loc:
        return "제주도"
    elif "대전" in loc:
        return "충청도"
    elif "인천" in loc:
        return "경기도"
    elif "상주" in loc:
        return "경상도"
    elif "광주광역시" in loc:
        return "전라도"
    else:
        return loc
```

```python
for i in range(len(predict2good)):
    try:
        predict2good['MNTN_LOC'][i] = distribute_region(predict2good['MNTN_LOC'][i])
    except:
        predict2good['MNTN_LOC'][i] = predict2good['MNTN_LOC'][i]
```
